```python
# column 이름 변경
electrical_energy.columns = ['year','month','day','installed_capacity','supply_capacity','full_power','supply_reserve','supply_reserve_ratio','standard_date']
```

```python
# 2004년 7월 이전에는 자료가 띄엄띄엄 존재해서 정리해줌
electrical_energy[electrical_energy['year']<2004]
```

```python
dels = electrical_energy[electrical_energy['year']<2004].index
electrical_energy.drop(dels, axis=0, inplace = True)
```

```python
electrical_energy = electrical_energy.sort_values(by=['year','month','day'],ascending=True).reset_index(drop=True)
electrical_energy
```

```python
electrical_energy.drop(electrical_energy.index[:3],inplace=True)
electrical_energy
```

```python
# 필요없는 column 제거
dels = ['supply_reserve','supply_reserve_ratio','standard_date']
electrical_energy.drop(dels, axis=1, inplace=True)
electrical_energy
```

```python
# 이후 기상데이터와 합치기 위해 datetime으로 타입 변경
electrical_energy['date'] = pd.to_datetime(electrical_energy[['year','month','day']])
electrical_energy
```

```python
# 필요없는 날짜 정보 제거
dels = ['year','month','day']
electrical_energy.drop(dels, axis=1, inplace=True)

# 열 정렬
electrical_energy = electrical_energy[['date', 'installed_capacity', 'supply_capacity', 'full_power']]
electrical_energy
```

```python
# 정리한 데이터 저장
electrical_energy.to_csv('electrical_energy.csv', index=False)
```
