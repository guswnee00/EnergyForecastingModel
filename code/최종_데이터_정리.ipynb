```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
# 파일 업로드
from google.colab import files
files.upload()
electrical_energy = pd.read_csv('electrical_energy.csv')
weather = pd.read_csv('weather.csv')
```

```python
# 잘 불러와졌는지 확인
weather.head()
```

```python
electrical_energy.head()
```

```python
# 전력 데이터에 맞춰서 기상 데이터도 2004/07/01 자료만 이용 -> date가 교집합인 데이터들 합치기
df = pd.merge(electrical_energy, weather, on ='date', how='inner')
df
```

```python
# 결측치 확인
df.isna().sum()
```

```python
# 중복값 확인
df.duplicated().sum()
```

```python
# 중복값 제거
df.drop_duplicates(keep='first',ignore_index = True, inplace=True)

# 확인
df.duplicated().sum()
```

```python
# date의 월 일만 가져오기 -> N234에서 year 특성을 포함할 경우 미래 사건을 예측하기 어려워져 삭제했다고 하길래..
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df
```

```python
# month 정보를 이용해 season 컬럼 추가
season = {12 : 'Winter', 1 : 'Winter', 2 : 'Winter', 
          3 : 'Spring', 4 : 'Spring', 5 : 'Spring',
          6 : 'Summer', 7 : 'Summer', 8 : 'Summer', 
          9 : 'Autumn', 10 : 'Autumn', 11 : 'Autumn'}
df['season'] = df[['month']].apply(lambda x : x.map(season))

# 확인
df.head()
```

```python
# date 컬럼 제거
df.drop('date', axis=1, inplace = True)

# 컬럼 순서 정리
df = df[['season', 'month', 'day', 'installed_capacity', 'supply_capacity', 
         'full_power','average_temperature', 'highest_temperature', 
         'lowest_temperature','precipitation', 'average_wind_speed', 
         'max_wind_speed', 'average_humidity', 'lowest_humidity']]

# 확인
df
```

```python
# 정리한 데이터 저장
df.to_csv('data.csv', index=False)
```
